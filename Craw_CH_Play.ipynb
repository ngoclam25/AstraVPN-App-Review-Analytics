```python
!pip install google-play-scraper
import pandas as pd
from google_play_scraper import Sort, reviews
import time

print("--- BẮT ĐẦU QUY TRÌNH CÀO DỮ LIỆU ĐA ỨNG DỤNG (COMPETITOR BENCHMARKING) ---")

apps_to_scrape = {
    'AstraVPN': 'com.astravpn.security.vpnproxy',
    '1.1.1.1 + WARP': 'com.cloudflare.onedotonedotonedotone',
    'NordVPN': 'com.nordvpn.android'
}

all_reviews_list = []

for app_name, app_id in apps_to_scrape.items():
    print(f"\n Đang cào dữ liệu cho app: {app_name} (ID: {app_id})...")

    try:
        raw_reviews, _ = reviews(
            app_id,
            lang='en',
            country='us',
            sort=Sort.NEWEST,
            count=3500,
            filter_score_with=None
        )


        df_app = pd.DataFrame(raw_reviews)

        if not df_app.empty:
            df_app = df_app[['userName', 'score', 'content', 'at', 'appVersion']]
            df_app['app_name'] = app_name
            all_reviews_list.append(df_app)
            print(f" Đã lấy được {len(df_app)} đánh giá của {app_name}.")
        else:
            print(f"Ko thấy đánh giá nào {app_name}.")

    except Exception as e:
        print(f"Lỗi khi cào app {app_name}: {e}")

    time.sleep(2)

print("\n--- GIAI ĐOẠN GỘP VÀ CHUẨN HÓA DỮ LIỆU ---")

if all_reviews_list:
    df_final = pd.concat(all_reviews_list, ignore_index=True)
    df_final.columns = ['user_name', 'score', 'content', 'review_date', 'app_version', 'app_name']
    df_final = df_final[['app_name', 'user_name', 'score', 'content', 'review_date', 'app_version']]
    output_filename = 'vpn_competitor_reviews_raw.csv'
    df_final.to_csv(output_filename, index=False, encoding='utf-8-sig')
```
